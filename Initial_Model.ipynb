Lets first import files that we saved after preprocessing part

```python
import pandas as pd
```

```python
x_train = pd.read_csv('X_train2.csv')
y_train = pd.read_csv('y_train.csv')
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_train.shape
```

```python
y_train.shape
```

Lets separate the damage grade from y_train since y_train also cotains index values

```python
damage_grade = y_train['damage_grade']
```

```python
damage_grade.shape
```

```python
x_test = pd.read_csv('X_test2.csv')
x_test.head()
```

We are using RandomForestClassifier to train our model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier()
clf.fit(x_train,damage_grade)
```

```python
building_id = x_test['building_id']
test = x_test.drop(columns=['building_id'])
```

```python
test.head()
```

```python
test.shape
```

lets run our model on test data to predict the damage grades

```python
pred = clf.predict(test)
```

Lets save the final result to output.csv file with the damage grade and building id specified

```python
final_result = pd.DataFrame({ 'building_id' : building_id , 'damage_grade' : pred})
final_result.to_csv('output.csv', index = False)
```

We can check the accuracy our predicted model by uploading this output file in hackerearth competition link as mentinoned: 
https://www.hackerearth.com/challenge/competitive/machine-learning-challenge-6-1/
Lets analyse the important features

```python
feature_imp = pd.Series(clf.feature_importances_,index=x_train.columns).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)
# Add labels to your graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.show()
```

So we can thus remove the features that are impacting our model to a very little extent in final_model file.
This can increase our accuracy
