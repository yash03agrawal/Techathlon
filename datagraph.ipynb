# CODE FOR GRAPH GENERATIONS

```python
import pandas as pd
import numpy as np
```

```python
#test csvfile
test_data1=pd.read_csv('Dataset/test.csv')

#building structure csvfile
building_structure=pd.read_csv('Dataset/building_structure.csv')

#output predicted file for the test data
final_result=pd.read_csv('output.csv')  
```

```python
print(test_data1.info())
```

```python
#raw combined data

x = pd.DataFrame()
x['building_id']=test_data1['building_id']
x['district_id']=test_data1['district_id']
x = pd.merge(x,building_structure,how='left',on='building_id')
x['floors_diff']=x['count_floors_pre_eq']-x['count_floors_post_eq']
x['mun_id']=test_data1['vdcmun_id']
x['grade']=final_result['damage_grade']
print(x)
x
```

```python
#sorting with respect to 'grade' and then wrt 'floors_diff'

y = x.sort_values(by=['grade','floors_diff'],ascending=False)
print(y)
```

```python
#district wise sorted val

z = y.sort_values(by=['district_id_x'],kind='mergesort')
z[:1000]
```

```python
#31 files for districts

files,f=[],[]
for i in range(31):
    files.append("file"+str(i+1)+".csv")
print(files)
```

```python
grp=z.groupby('district_id_x')
counter=0
for name,group in grp:
    print(name)
    print(group)
    group.to_csv(files[counter], index = False)
    counter+=1
    if counter==31:
        break
```

```python
#graded wise distributed values

grading=z.drop(z.columns[1:-1],axis=1)
grading= pd.merge(grading,building_structure,how='left',on='building_id')
grading
```

```python
#considered the grade 1,5

grp1=grading.groupby('grade')
counter=0
for name,group in grp1:
    if name=='Grade 5':
        group.to_csv('grade_5.csv', index = False)
    if name=='Grade 1':
        group.to_csv('grade_1.csv', index = False)
    counter+=1
    if counter==5:
        break
```
